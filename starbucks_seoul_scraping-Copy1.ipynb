```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
starseoul = []

for shop in shops:
    name = shop.select('strong')[0].text.strip()
    add = shop.select('p')[0].text.strip()
    lat = str(shop['data-lat'])
    lon = str(shop['data-long'])
#     print(name,add,lat,lon)
    a = [name,add,lat,lon]
    starseoul.append(a)
    
starseoul
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(starseoul, columns=['name','address','latitude','longitude'])
pd_data
```

```python
pd_data['newaddress'] = pd_data['address'].str[:-9]
pd_data['phone'] = pd_data['address'].str[-9:]
pd_data = pd_data.drop(['address'], axis=1)
pd_data
```

```python
pd_data.rename(columns = {'newaddress' : 'address'}, inplace = True)
pd_data
```

```python
pd_data = pd_data[['name','address','phone','latitude','longitude']]
pd_data
```
