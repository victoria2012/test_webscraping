```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('https://search.shopping.naver.com/search/all?delivery=1&frm=NVSHATT&origQuery=%ED%82%A4%EB%B3%B4%EB%93%9C%20%EB%A7%88%EC%9A%B0%EC%8A%A4%20%EC%84%B8%ED%8A%B8&pagingIndex=1&pagingSize=40&productSet=total&query=%ED%82%A4%EB%B3%B4%EB%93%9C%20%EB%A7%88%EC%9A%B0%EC%8A%A4%20%EC%84%B8%ED%8A%B8&sort=rel&spec=M10015629%7CM10784864&timestamp=&viewType=list#')
```

```python
html = browser.page_source
html
```
