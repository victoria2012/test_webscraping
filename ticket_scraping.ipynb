```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
```

```python
driver.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
html = driver.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
len(soup), type(soup)
```

```python
tags = soup.select('a.prdName')
```

```python
len(tags), type(tags)
```

```python
tag = tags[0]
tag
```

```python
tag.text.strip()
```

```python
title = []
for tag in tags:
#     print(tag.text.strip())
    a = tag.text.strip()
    title.append(a)

title
```

```python
tags2 = soup.select('a.prdDuration')
tags2
```

```python
len(tags2), type(tags2)
```

```python
tag2 = tags2[0]
tag2
```

```python
tag2.text.strip()
```

```python
date = []
for tag2 in tags2:
#     print(tag2.text.strip())
    b = tag2.text.strip()
    date.append(b)
    
date
```

```python
result = list(zip(title,date))
result
```
