```python
result = list()

for team in teams:
    rank = team.select('strong')[0]
    name = team.select('span')[0]
    match = team.select('td:nth-child(3)')[0]
    win = team.select('td:nth-child(4)')[0]
    lose = team.select('td:nth-child(5)')[0]
    tie = team.select('td:nth-child(6)')[0]
    winrate = team.select('td:nth-child(7)')[0]
    g_behind = team.select('td:nth-child(8)')[0]
    con = team.select('td:nth-child(9)')[0]
    onbase = team.select('td:nth-child(10)')[0]
    slugging = team.select('td:nth-child(11)')[0]
    ten_match = team.select('td:nth-child(12)')[0]
#     print(rank.text.strip(), name.text.strip(), match.text.strip(), win.text.strip(), lose.text.strip(), tie.text.strip(), winrate.text.strip(), g_behind.text.strip(), con.text.strip(), onbase.text.strip(), slugging.text.strip(), ten_match.text.strip())
    a = [rank.text.strip(), name.text.strip(), match.text.strip(), win.text.strip(), lose.text.strip(), tie.text.strip(), winrate.text.strip(), g_behind.text.strip(), con.text.strip(), onbase.text.strip(), slugging.text.strip(), ten_match.text.strip()]
    result.append(a)
    
result
```

```python
import pandas as pd
```

```python
pd_data = pd.DataFrame(result, columns = ['Ranking','Name','Match','Win','Lose','Tie','Winning rate','Game behind','Continuation','On-base percentage','Slugging average','The last ten games'])
pd_data
```

```python
pd_data.to_excel('./saves/kbo_record_scraping.xls', index=False)
```
