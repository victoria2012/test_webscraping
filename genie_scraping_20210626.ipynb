```python
driver = webdriver.Chrome('./chromedriver.exe')
```

```python
driver.get('https://www.genie.co.kr/chart/top200')
```

```python
html = driver.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
songs = soup.select('td.info')
songs
```

```python
len(songs),type(songs)
```

```python
tag = songs[0]
```

```python
곡명 : a.title , 가수이름 : a.artist
```

```python
title = tag.select('a.title')
title
```

```python
len(title), type(title)
```

```python
artist = tag.select('a.artist')
artist
```

```python
genie = []
for tag in songs:
    title = tag.select('a.title')
    artist = tag.select('a.artist')
#     print(title[0].text.strip(), artist[0].text.strip())
    song = [title[0].text.strip(), artist[0].text.strip()]
    genie.append(song)
    
genie
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(genie, columns = ['Title','Artist'])
data
```

```python
data.to_excel('./saves/genie_20210626.xls', index=False)
```
